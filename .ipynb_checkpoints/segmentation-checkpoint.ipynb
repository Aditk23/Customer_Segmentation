```python
# Creating dummy variables

all_data=pd.get_dummies(all_data,columns=['Profession','Var_1','Spending_Score'])
```

```python
# Collecting training data

traindata = all_data[all_data.train_test==1]
```

```python
# Splitting training data into inputs and targets

X_train = traindata.drop(['Segmentation','train_test','ID'],axis=1)
y_train = traindata['Segmentation']
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Mapping target variable

y_train = y_train.map({'A':0,'B':1,'C':2,'D':3})
```

```python
y_train
```

```python
# Collecting testing data

testdata = all_data[all_data.train_test==0].drop(['train_test','ID'],axis=1)
```

```python
testdata.shape
```

```python
# As this dataset is from a cometition on AnalyticsVidhya we are not given target variable for test data

X_test = testdata.drop('Segmentation',axis=1)
```

```python
# Scaling data

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.fit_transform(X_test)
```

```python
# Implementing KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 1)
model.fit(X_train_scaled, y_train)
predictions = model.predict(X_test_scaled)
```

```python
predictions
```

```python
# Combining testing and training data with same ID's

ans = test.drop('Segmentation',axis=1)

ans['Segmentation'] = predictions

answer = ans[['ID','Segmentation']]

answer.Segmentation = answer.Segmentation.map({0:'A',1:'B',2:'C',3:'D'})

test1 = pd.read_csv('test.csv')
```

```python
target_filled = pd.merge(test1 , answer,   on='ID', how ='left' )
target_filled
```
